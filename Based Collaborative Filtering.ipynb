```python
"""
Going to build an Item based Collaborative Recommendation System. 
Import all the necessary packages

Item based, focuses mainly on the relationships between items and not on people. 
"""

import numpy as np # fast computation
import pandas as pd # for data analysis. 
import matplotlib.pyplot as plt # for Visualization

```

```python
#load the dataset to Pandas DataFrame 

r_cols  = ["user_id","movie_id", "rating"]

#take the first 3 columns 
user_ratings = pd.read_csv("/Users/admin/Desktop/ml-100k/u.data",sep="\t",names=r_cols,usecols=range(3),encoding="ISO-8859-1")
```

```python
#get the Movie id and `title of the Movie

# columns for the movie and Id
m_cols = ["movie_id", "title"]
movies = pd.read_csv("/Users/admin/Desktop/ml-100k/u.item",sep="|",names=m_cols,usecols=range(2), encoding="ISO-8859-1")
```

```python
# merge the two Dataframe
ratings = pd.merge(movies,user_ratings,on="movie_id")
ratings.head()
```

```python
# Now goign use pandas pivot table to construct a new Dataframe. That show which movies were rated by a user 
# I think the NAN means the user did not watch or rate the 

movieRatings = ratings.pivot_table(index=["user_id"],columns=["title"],values="rating")

movieRatings.head(10)
```

```python
#checking if the Dataframe makes sense based popular movies

dalmations = movieRatings["101 Dalmatians (1996)"]
dalmations.head(20) # this wasn't a liked movie
```

```python
starWars = movieRatings["Star Wars (1977)"]
starWars.head(20) # alot of people like star Wars 
```

```python
# Since some movies have low rating, NAN and some has High rating
# I want to find the connection between two or more things using the Pandas correlation function 
# using the Star Wars movie ratings

same_movies = movieRatings.corrwith(starWars) # gives the correlation between the movies. 

same_movies.head()
# shows the correlation score. 
```

```python
# Now going all the Missing results (NAN )
same_movies = same_movies.dropna()

dataFrame = pd.DataFrame(same_movies) # loads to a new Df 

dataFrame.head()
```

```python
same_movies.sort_values(ascending=True)
```

```python
# These results are quite skeptic
#Movies I never heard of, correlates with Star Wars ? 
# I think a few people has watched some of these movies and gave good ratings
# We cannot recommend those movies to every one else

```

```python
# this gets the movie titles, total ratings for each movie and the average(mean) of the ratings. 
# Uses the ratings DataFrame. 
movieStats = ratings.groupby("title").agg({"rating":[np.size, np.mean]})

movieStats.head(50) 
#prints out the stats 
```

```python
#I will remove movies that are rated less then 100 people 
# too see which movies are really good, and to recommend. 

#don't take any rating less than 100. 
popularMovies = movieStats["rating"]["size"] >=100 # 100 is my threshold for this dataset. 
popularMovies 

#there are a few actual good movies in this dataset that have been rated by more than 100 people 
```

```python
movieStats[popularMovies].sort_values([("rating","mean")], ascending=False)[:15]
#These movies seem more good. 
```

```python
# Now going to Join this data with our original set of similar movies to star Wars
#because star Wars has the highest Rating. 

dataFrame = movieStats[popularMovies].join(pd.DataFrame(same_movies, columns=["similarity"]))

dataFrame.sort_values(["similarity"],ascending=False)[:10]
```
